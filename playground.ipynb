# Put and call options

```python
x = np.arange(0., 100, 0.01)
```

```python
strike_price_call = 50
strike_price_put = 50
cost_option = 10
CV = np.maximum(x, strike_price_call) - strike_price_call
PV = strike_price_put - np.minimum(x, strike_price_put)
CPL = CV - cost_option*2
PPL = PV - cost_option*2
plt.figure(figsize=(10,5))
plt.plot(x, CV, label='Call Value')
plt.plot(x, PV, label='Put Value')
plt.plot(x, CPL, label='Call Profit/Loss')
plt.plot(x, PPL, label='Put Profit/Loss')
plt.title('cost of option=' + str(cost_option) + ', strike price call=' + str(strike_price_call) + ', strike price put=' + str(strike_price_put))
plt.legend()
plt.xlabel('Stock Value')
plt.grid()
plt.show()
```

# Geithner Plan

```python
x = np.arange(0., 100, 5)
```

```python
I = 30 # initial stock value
r = .15 # percentage invested in the toxic asset from a hedge fund
S = I*r # money invested in the toxic asset from a hedge fund
N = 2 # numbers of shares bought 
E = 2*S # total equity
L = int(N*I - E) # loan 
T = N*I # assets
Gplan = np.maximum(I-S + (N*x-L)/2, I-S)
strike_price_put = I - S
cost_option = 0
PV = strike_price_put - np.minimum(x, strike_price_put)

plt.figure(figsize=(10,5))
plt.plot(x, x, label='Stock Value')
plt.plot(x, Gplan, '*', label='Hedge Fund + Geithner Plan')
plt.plot(x, PV, label='Equivalent Put Value')
plt.plot(x, PV + x, '--', label='Stock + Equivalent Put Value')
plt.title('N=' + str(N) + ', L=' + str(L) + ', I=' + str(I) + ', r=' + str(r) + ', S=' + str(S) + ', strike price of equivalent put=' + str(strike_price_put))
plt.legend()
plt.xlabel('Stock Value')
plt.grid()
plt.show()
```
